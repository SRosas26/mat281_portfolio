El conjunto de datos, denominado `company_sales_data.csv`, recoge las ventas mensuales de distintos productos de una empresa, ofreciendo una visión general de las unidades vendidas y los beneficios generados. Entre los productos incluidos se encuentran cremas faciales, jabones de baño y artículos para el cuidado capilar. Este conjunto de datos es útil para analizar el rendimiento de ventas y el comportamiento del mercado a lo largo del tiempo.

### Descripción de los Datos:

1. **month_number**: Número correspondiente al mes del año (1 para enero, 2 para febrero, etc.).
2. **facecream**: Cantidad de unidades de crema facial vendidas durante el mes.
3. **facewash**: Cantidad de unidades de limpiador facial vendidas durante el mes.
4. **toothpaste**: Cantidad de unidades de pasta de dientes vendidas durante el mes.
5. **bathingsoap**: Cantidad de unidades de jabón de baño vendidas durante el mes.
6. **shampoo**: Cantidad de unidades de champú vendidas durante el mes.
7. **moisturizer**: Cantidad de unidades de crema hidratante vendidas durante el mes.
8. **total_units**: Total de unidades vendidas de todos los productos en un mes dado.
9. **total_profit**: Ganancia total obtenida por la venta de todos los productos durante el mes, expresada en la moneda correspondiente.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2024/main/docs/labs/data/company_sales_data.csv")
df.head()
```

1. Lea la columna **`total_profit`** correspondiente a todos los meses y represente los datos utilizando un gráfico lineal y un gráfico de dispersión.


```python
#Representacion en grafico lineal del total_profit para todos los meses

#definimos los datos para los ejes x e y
x = df['month_number']
y = df['total_profit']

#crear grafico lineal
plt.plot(x, y)
plt.xlabel('Meses')
plt.ylabel('Total Profit')
plt.title('Ganancia total por Mes')
plt.show()
```

```python
#crear grafico de dispersion
plt.scatter(x, y)
plt.xlabel('Meses')
plt.ylabel('Total Profit')
plt.title('Ganancia total por Mes')
plt.show()
```

**Conclusion**:
se puede apreciar que las ganancias por mes, se pueden apreciar mejor en un grafico lineal que en uno por dispersion, ya que el lineal ayuda a visualizar mejor las fluctuaciones de las ganancias por mes.

2. Lea los datos de ventas de todos los productos y represéntelos en un gráfico de líneas múltiples.


```python
# Crear el gráfico de líneas múltiples para las ventas de todos los productos
plt.figure(figsize=(10, 6))
# Use the index of the DataFrame for the x-axis
plt.plot(df.index, df['facecream'], label='Face Cream', marker='o')
plt.plot(df.index, df['facewash'], label='Face Wash', marker='o')
plt.plot(df.index, df['toothpaste'], label='Toothpaste', marker='o')
plt.plot(df.index, df['bathingsoap'], label='Bathing Soap', marker='o')
plt.plot(df.index, df['shampoo'], label='Shampoo', marker='o')
plt.plot(df.index, df['moisturizer'], label='Moisturizer', marker='o')

plt.title('Ventas de todos los productos por mes')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.legend()
plt.grid(True)
plt.show()

```

Conclusion: Se observa que Toothpaste tiene un crecimiento sostenido y es el producto con mayores ventas a lo largo de los meses, seguido de Bathing Soap y Face Cream, que también muestran un aumento progresivo pero en menor proporción. Shampoo, Moisturizer y Face Wash tienen un crecimiento más moderado en comparación. La tendencia general indica que todos los productos experimentan un aumento en ventas, pero con diferentes ritmos, siendo los productos de higiene personal más básicos (como pasta de diente y jabón) los que presentan un mayor crecimiento.

3. Lea los datos de ventas de **`facecream`** y **`facewash`** y represéntelos utilizando un gráfico de barras.

```python
# Grafico de barras las ventas de facecream y facewash
x = ['facecream','facewash']
y = [df['facecream'].sum(),df['facewash'].sum()]
plt.bar(x, y)
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.title('Ventas de Facecream y Facewash')
plt.show()
```

**Conclusion:**
Las ventas de facecream son mucho mayores a las ventas de facewash.

4. Lea los datos de ventas de todos los productos y represéntelos utilizando un gráfico de caja (*box-plot*).


```python
#leer datos de venta de todos los productos
productos = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
ventas = df[productos].values
#representarlos utilizando un grafico de caja (box-plot)
plt.boxplot(ventas, labels=productos)
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.title('Ventas de todos los Productos')
plt.show()
```

**Conclusion:**
La gráfica muestra que el Bathing Soap es el producto con mayores ventas y la mayor dispersión, seguido por Toothpaste, que también tiene un buen desempeño. Por otro lado, productos como Face Wash y Moisturizer registran las ventas más bajas y con menor variabilidad. Esto sugiere que los productos de higiene básica como jabones y pastas dentales tienen una demanda significativamente mayor en comparación con los productos de cuidado personal más específicos.

5. Calcule las ventas totales del año pasado para cada producto y represéntelas utilizando un gráfico circular.

```python
plt.figure(figsize=(10,6))
total_ventas = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
plt.pie(total_ventas, labels=total_ventas.index, autopct='%1.1f%%')
plt.title('Ventas Totales por Producto')
plt.show()
```

Conclusion: Podemos observar que Bathing Soap domina con el 40.6% de las ventas totales, seguido por Toothpaste, que representa el 24.9%. Face Cream ocupa el tercer lugar con el 12.3%, mientras que Shampoo se queda con el 9.0%. Tanto Face Wash como Moisturizer tienen la menor participación, cada uno con un 6.6%. Esto muestra claramente que los productos de higiene básica como el jabón y la pasta de diente son los más vendidos, mientras que los productos más especializados tienen una demanda significativamente menor.

**Conclusión General**: Se concluye que los productos de higiene básica, como Bathing Soap y Toothpaste, son los más vendidos, representando más del 65% de las ventas totales. Bathing Soap es el líder destacado con más del 40% del total, seguido de Toothpaste con cerca del 25%. Los productos especializados como Face Cream, Shampoo, Face Wash y Moisturizer tienen una participación menor, siendo Face Cream el más destacado de este grupo con un 12.3%. A lo largo del año, las ganancias muestran peaks en meses específicos, lo que sugiere fluctuaciones en la demanda, posiblemente relacionadas con factores estacionales o eventos promocionales. En general, los productos de uso diario superan ampliamente a los productos más especializados en términos de demanda, lo que refleja una preferencia del mercado por los artículos de higiene básica.
