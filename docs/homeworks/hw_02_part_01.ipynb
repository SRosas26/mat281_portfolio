```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
districts_info = pd.read_csv('drive/MyDrive/COVID19/districts_info.csv')
products_info = pd.read_csv('drive/MyDrive/COVID19/products_info.csv')
engagement_data = pd.read_csv('drive/MyDrive/COVID19/1142.csv')
```

```python
# Configuración de gráficos
sns.set(style="whitegrid")

# Limpieza de la columna de conectividad
districts_info['county_connections_ratio'] = districts_info['county_connections_ratio'].apply(
    lambda x: float(str(x).replace('[', '').replace(']', '').split(',')[0]) if pd.notnull(x) else None
)

# Gráfico de la distribución de la conectividad del condado
plt.figure(figsize=(12, 6))
sns.histplot(districts_info['county_connections_ratio'], bins=20, kde=True, color='blue')
plt.title('Distribución de la Relación de Conectividad del Condado')
plt.xlabel('Relación de Conectividad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Mostrar estadísticas descriptivas
print("Estadísticas descriptivas de la relación de conectividad:")
print(districts_info['county_connections_ratio'].describe())
```

```python

# Configuración de gráficos
sns.set(style="whitegrid")

# Convertir las fechas
engagement_data['time'] = pd.to_datetime(engagement_data['time'])

# Filtrar datos para el año 2020
engagement_data_2020 = engagement_data[engagement_data['time'].dt.year == 2020]

# Calcular la participación promedio por mes en 2020
engagement_data_2020['month'] = engagement_data_2020['time'].dt.month
avg_monthly_engagement = engagement_data_2020.groupby('month')['engagement_index'].mean()

# Gráfico de la participación digital promedio por mes en 2020
plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_monthly_engagement.index, y=avg_monthly_engagement.values, marker='o', color='green')
plt.title('Índice de Participación Promedio por Mes en 2020')
plt.xlabel('Mes')
plt.ylabel('Índice de Participación Promedio')
plt.grid(True)
plt.show()

# Mostrar estadísticas descriptivas
print("Estadísticas descriptivas del índice de participación en 2020:")
print(engagement_data_2020['engagement_index'].describe())
```


La relación de conectividad en los distritos muestra que la mayoría de los condados tienen una conectividad baja y homogénea, centrada en un valor de aproximadamente 0.18, lo que sugiere limitaciones en la infraestructura de internet. Por otro lado, el índice de participación digital en 2020 presenta un comportamiento variable, con un notable aumento en los primeros meses del año, alcanzando su punto máximo en marzo-abril, coincidiendo con el inicio de la pandemia de COVID-19. Posteriormente, se observa una caída durante los meses de verano y una recuperación en otoño con el regreso a clases. Este análisis conjunto resalta la importancia de mejorar la infraestructura de conectividad para mantener un acceso equitativo y un compromiso digital constante, especialmente en situaciones de emergencia como la pandemia.

```python


# Configuración de gráficos
sns.set(style="whitegrid")

# Convertir las fechas
engagement_data['time'] = pd.to_datetime(engagement_data['time'])

# Filtrar datos para el año 2020
engagement_data_2020 = engagement_data[engagement_data['time'].dt.year == 2020]

# Dividir el año en trimestres
engagement_data_2020['quarter'] = engagement_data_2020['time'].dt.to_period('Q')

# Calcular la participación promedio por trimestre en 2020
avg_quarterly_engagement = engagement_data_2020.groupby('quarter')['engagement_index'].mean()

# Gráfico de la participación digital promedio por trimestre en 2020
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_quarterly_engagement.index.astype(str), y=avg_quarterly_engagement.values, palette='viridis')
plt.title('Índice de Participación Promedio por Trimestre en 2020')
plt.xlabel('Trimestre')
plt.ylabel('Índice de Participación Promedio')
plt.grid(True)
plt.show()

# Mostrar estadísticas descriptivas
print("Estadísticas descriptivas del índice de participación por trimestre en 2020:")
print(avg_quarterly_engagement.describe())
```


El análisis del índice de participación digital por trimestre en 2020 muestra que el primer trimestre (Q1) tuvo un nivel alto de participación, probablemente debido al inicio de la pandemia de COVID-19 y al cierre de las escuelas, lo que forzó a los estudiantes y docentes a adaptarse rápidamente al aprendizaje en línea. En el segundo trimestre (Q2), el índice se mantuvo relativamente alto, reflejando una continuación del uso de plataformas digitales mientras la pandemia avanzaba. Sin embargo, en el tercer trimestre (Q3), se observa una disminución notable en la participación, que podría asociarse con las vacaciones de verano y la adaptación a nuevas circunstancias educativas. Finalmente, en el cuarto trimestre (Q4), la participación aumentó nuevamente, posiblemente debido al regreso a clases y una mejor preparación de las instituciones para afrontar el aprendizaje a distancia. Este comportamiento sugiere que la pandemia tuvo un impacto significativo en la forma en que se llevó a cabo la educación, y el índice de participación digital refleja cómo las instituciones y los estudiantes se adaptaron a lo largo del año.

```python

# Configuración de gráficos
sns.set(style="whitegrid")

# Convertir las fechas en engagement_data
engagement_data['time'] = pd.to_datetime(engagement_data['time'])

# Filtrar datos para el año 2020
engagement_data_2020 = engagement_data[engagement_data['time'].dt.year == 2020]

# Unir engagement_data con products_info para obtener el tipo de producto
merged_data = pd.merge(engagement_data_2020, products_info, left_on='lp_id', right_on='LP ID', how='inner')

# Calcular la participación promedio por tipo de tecnología educativa
merged_data['quarter'] = merged_data['time'].dt.to_period('Q')
avg_engagement_by_type = merged_data.groupby(['Primary Essential Function'])['engagement_index'].mean()

# Seleccionar las 5 tecnologías más utilizadas
top_5_types = avg_engagement_by_type.nlargest(5).index
filtered_data = merged_data[merged_data['Primary Essential Function'].isin(top_5_types)]
avg_engagement_top_5 = filtered_data.groupby(['quarter', 'Primary Essential Function'])['engagement_index'].mean().unstack()

# Gráfico de la participación promedio por tipo de tecnología educativa (Top 5)
plt.figure(figsize=(14, 8))
avg_engagement_top_5.plot(kind='line', marker='o')
plt.title('Participación Promedio por Tipo de Tecnología Educativa (Top 5) durante 2020')
plt.xlabel('Trimestre')
plt.ylabel('Índice de Participación Promedio')
plt.legend(title='Tipo de Tecnología Educativa', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

El gráfico de las 5 tecnologías educativas más utilizadas en 2020 muestra que SDO - Learning Management Systems (LMS) lideró la participación promedio a lo largo del año, con una caída en los trimestres Q2 y Q3 y un aumento en Q4, indicando su consolidación en la enseñanza a distancia. CM - Virtual Classroom mantuvo un uso constante, reflejando la necesidad de interacción en tiempo real durante la educación remota. Las tecnologías LC - Content Creation & Curation y LC - Study Tools también registraron incrementos en los últimos dos trimestres, sugiriendo que docentes y estudiantes comenzaron a utilizar más herramientas de creación de contenido y apoyo al estudio para enriquecer el proceso educativo. LC - Sites, Resources & Reference mostró un comportamiento similar, evidenciando un mayor uso conforme avanzaba el año, lo que apunta a una diversificación de los recursos digitales para fortalecer la educación remota y adaptarse a las nuevas dinámicas de aprendizaje.

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de los gráficos
sns.set_theme(style="whitegrid")
plt.style.use('default')

# Unir los datos de participación con los de distritos
engagement_districts = engagement_data.merge(districts_info, left_index=True, right_index=True, how='inner')

# Gráfico de barras de acceso promedio a plataformas por tipo de localidad
plt.figure(figsize=(12, 6))
sns.barplot(data=engagement_districts, x='locale', y='pct_access', ci=None, palette="Set2")
plt.title('Acceso Promedio a las Plataformas por Tipo de Localidad')
plt.xlabel('Tipo de Localidad')
plt.ylabel('Porcentaje de Acceso Promedio')
plt.show()

# Gráfico de dispersión del índice de compromiso por porcentaje de estudiantes negros/hispanos
plt.figure(figsize=(12, 6))
sns.scatterplot(data=engagement_districts, x='pct_black/hispanic', y='engagement_index', hue='locale', palette="Set1", alpha=0.6)
plt.title('Índice de Compromiso por Porcentaje de Estudiantes Negros/Hispanos')
plt.xlabel('Porcentaje de Estudiantes Negros/Hispanos')
plt.ylabel('Índice de Compromiso')
plt.legend(title='Tipo de Localidad')
plt.show()

# Gráfico de barras del índice de compromiso promedio por inversión por estudiante
plt.figure(figsize=(14, 6))
sns.barplot(data=engagement_districts, x='pp_total_raw', y='engagement_index', ci=None, palette="cool")
plt.title('Índice de Compromiso Promedio por Inversión por Estudiante')
plt.xlabel('Inversión por Estudiante')
plt.ylabel('Índice de Compromiso Promedio')
plt.xticks(rotation=45)
plt.show()
```


Los gráficos revelan que la participación de los estudiantes en plataformas de aprendizaje en línea está influenciada por factores geográficos, demográficos y socioeconómicos. Los distritos rurales muestran un acceso promedio más alto a las plataformas en comparación con los suburbios, ciudades y pueblos. En el contexto demográfico, el índice de compromiso varía significativamente en los distritos con un menor porcentaje de estudiantes negros/hispanos, observándose menor compromiso en áreas con mayor diversidad. En cuanto al estatus socioeconómico, el análisis sugiere que los distritos con menor inversión por estudiante muestran un mayor índice de compromiso promedio, mientras que los distritos con mayor inversión tienen un compromiso más bajo, lo que puede indicar una relación inversa entre inversión y compromiso. En resumen, la participación y el compromiso están influenciados por la geografía y la demografía, y sorprendentemente, un menor estatus socioeconómico parece correlacionarse con un mayor compromiso de los estudiantes.

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos (reemplaza 'ruta/a/tu/archivo.csv' con la ubicación de tu archivo)
student_data = pd.read_csv('drive/MyDrive/COVID19/1142.csv')

# Convertir la columna 'time' a formato de fecha
student_data['time'] = pd.to_datetime(student_data['time'])

# Calcular el promedio diario de acceso y compromiso
daily_trends = student_data.groupby('time').mean().reset_index()

# Gráfico de tendencia temporal del acceso promedio
plt.figure(figsize=(14, 6))
plt.plot(daily_trends['time'], daily_trends['pct_access'], label='Promedio de Acceso (%)', color='blue')
plt.title('Tendencia Temporal del Acceso Promedio a las Plataformas')
plt.xlabel('Fecha')
plt.ylabel('Porcentaje de Acceso Promedio')
plt.grid(True)
plt.legend()
plt.show()

# Gráfico de tendencia temporal del índice de compromiso promedio
plt.figure(figsize=(14, 6))
plt.plot(daily_trends['time'], daily_trends['engagement_index'], label='Índice de Compromiso Promedio', color='green')
plt.title('Tendencia Temporal del Índice de Compromiso Promedio')
plt.xlabel('Fecha')
plt.ylabel('Índice de Compromiso Promedio')
plt.grid(True)
plt.legend()
plt.show()
```

Los gráficos de tendencias temporales muestran cómo el acceso promedio a las plataformas de aprendizaje y el índice de compromiso de los estudiantes evolucionaron durante el año 2020. En el primer gráfico, se observa un patrón fluctuante en el porcentaje de acceso, con picos a inicios de 2020 y una notable disminución durante los meses de verano (mediados de 2020), posiblemente relacionada con las vacaciones o ajustes en las políticas escolares. Hacia el final del año, el acceso vuelve a aumentar, lo que podría coincidir con el retorno a clases o el refuerzo de medidas educativas. En el segundo gráfico, el índice de compromiso muestra un comportamiento más variable, con un pico muy pronunciado en mayo, lo que podría corresponder a un evento o intervención significativa en esa fecha. A partir de mediados de año, el compromiso se estabiliza en niveles más bajos, lo que sugiere que los estudiantes pudieron haber experimentado una menor motivación o cambios en la modalidad de aprendizaje.
